```python
import cv2
import matplotlib.pyplot as plt
```

```python
demo1_path = './demo/IMG_2782.JPG'
demo1 = cv2.imread(demo1_path)
demo2_path = './demo/IMG_2887.JPG'
demo2 = cv2.imread(demo2_path)
```

# show demo image

```python
demo1_rgb = cv2.cvtColor(demo1, cv2.COLOR_BGR2RGB)
demo2_rgb = cv2.cvtColor(demo2, cv2.COLOR_BGR2RGB)

print("demo1")
plt.imshow(demo1_rgb)
plt.show()

print("\n\n")

print("demo2")
plt.imshow(demo2_rgb)
plt.show()
```

# cut top and bottom

```python
noise_t=1

y1=edges1.sum(axis=1)
x1=edges1.sum(axis=0)
x_len1 = len(x1)
y_len1 = len(y1)
x_middle1 = int(x_len1/2)
y_middle1 = int(y_len1/2)

# non-plate rows
y_0_indices1 = [i for i, val in enumerate(y1) if val <noise_t]
y_top_list1 = [top for top in y_0_indices1 if top < y_middle1]

# top edges
if(len(y_top_list1)!=0):
    y_top1 = y_top_list1[-1]
else:
    y_top1 = 0
    
# bottom edges
y_bottom_list1 = [bottom for bottom in y_0_indices1 if bottom > y_middle1]
if(len(y_bottom_list1)!=0):
    y_bottom1 = y_bottom_list1[0]
else:
    y_bottom1 = y_len1


y2=edges2.sum(axis=1)
x2=edges2.sum(axis=0)
x_len2 = len(x2)
y_len2 = len(y2)
x_middle2 = int(x_len2/2)
y_middle2 = int(y_len2/2)

# non-plate rows
y_0_indices2 = [i for i, val in enumerate(y2) if val <noise_t]
y_top_list2 = [top for top in y_0_indices2 if top < y_middle2]

# top edges
if(len(y_top_list2)!=0):
    y_top2 = y_top_list2[-1]
else:
    y_top2 = 0

# bottom edges
y_bottom_list2 = [bottom for bottom in y_0_indices2 if bottom > y_middle2]
if(len(y_bottom_list2)!=0):
    y_bottom2 = y_bottom_list2[0]
else:
    y_bottom2 = y_len2


print("noise_t = %d"%noise_t)

print("demo1_2 non-plate rows")
plt.imshow(edges1)
for y_0_index1 in y_0_indices1:
    plt.plot([0, x_len1], [y_0_index1, y_0_index1], c='blue')
plt.show()

print("demo2_2 non-plate rows")
plt.imshow(edges2)
for y_0_index2 in y_0_indices2:
    plt.plot([0, x_len2], [y_0_index2, y_0_index2], c='blue')
plt.show()
```
